# Create Custom Features

```python
# Create Custom Feature: Adjusted Total Basement Area
df['AdjTotalBsmtSF'] = df['BsmtFinSF1']+df['BsmtFinSF2']+0.9*df['BsmtUnfSF']
```

```python
# Create Custom Feature: Adjusted Total Bathroom 
df['AdjTotalBath'] = df['FullBath']+df['BsmtFullBath']+0.6*(df['HalfBath']+df['BsmtHalfBath'])
```

```python
# Create Custom Feature: HouseAge
df['HouseAge'] = df['YrSold']-(0.6*df['YearRemodAdd']+0.4*df['YearBuilt'])
```

```python
# Transform SalePrice to LogSalePrice 
df['LogSalePrice'] = np.log(df['SalePrice'])
```

```python
df['AdjOutdoorSF'] = df['3SsnPorch']+df['EnclosedPorch']+df['OpenPorchSF']+df['ScreenPorch']+df['WoodDeckSF']
```

```python
df.shape
```

# Helper Functions

```python
def indices_of_outliers(df, column_name, stdev_threshold):
    mask_of_outliers = np.abs(df[column_name] - df[column_name].mean()) >= (stdev_threshold*df[column_name].std())
    set_of_indices_of_outliers = df.index[mask_of_outliers].tolist()
    return set_of_indices_of_outliers

# example
indices_of_outliers(df, "LotFrontage", 4)
```

```python
def remove_rows_with_outliers(df, list_of_column_names, stdev_threshold):
    outlier_index = []
    for col in list_of_column_names:
        set_of_indices_of_outliers = indices_of_outliers(df, col, stdev_threshold)
        outlier_index.extend(set_of_indices_of_outliers)
    outlier_index = list(set(outlier_index))
    df_without_outliers = df.drop(df.index[outlier_index])
    return df_without_outliers

# example
list_of_column_names = ["LotFrontage", "OverallQual", "LotArea"]
remove_rows_with_outliers(df, list_of_column_names, stdev_threshold = 4)
```

# NA Imputation

```python
# check NA in original df
df.isnull().sum()[df.isnull().sum()>0]
```

```python
series_names = ['BsmtExposure','BsmtQual','FireplaceQu','GarageQual','GarageType','GarageYrBlt','GarageFinish','GarageCond','PoolQC','Fence','MiscFeature','Alley','BsmtCond','BsmtFinType1','BsmtFinType2']
for series_name in series_names:
    df.loc[df[series_name].isnull(),series_name] = 'None'
    print(df[series_name].value_counts(sort = True))
    print('*'*20)
```

```python
df.isnull().sum()[df.isnull().sum()>0]
```

```python
# impute using neighborhood (just for now, might change later)
# df['MasVnrType'] = df.groupby(['Neighborhood'])['MasVnrType'].transform(lambda x: x.fillna(x.mode()[0]))
df['LotFrontage'] = df.groupby(['Neighborhood'])['LotFrontage'].transform(lambda x: x.fillna(x.median()))
df['Electrical'] = df['Electrical'].fillna('SBrkr')
```

```python
df.isnull().sum()[df.isnull().sum()>0]
```

### Converting Categorical Features 

```python
df.GarageYrBlt.dtype
```

```python
df.select_dtypes(include=[object])
```

```python
df_categorical = df.select_dtypes(include=[object]).drop(['MasVnrType','GarageYrBlt'], axis = 1) # 43 categorical columns
categorical_columns = df_categorical.columns.to_list()
df_categorical.sample(5)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_categorical_le = df_categorical.apply(le.fit_transform)
df_categorical_le.head(5)
```

### Combining Label Encoded Categorical Columns and Numeric Columns

```python
df_numeric = df._get_numeric_data().drop('MasVnrArea',axis = 1) # need to impute missing value for MasVnrArea
numeric_columns = df_numeric.columns.to_list()
```

```python
df_le = pd.DataFrame(np.concatenate([df_numeric,df_categorical_le],axis = 1))
df_le.columns = numeric_columns + categorical_columns
df_le.head()
```

```python
X = df_le.drop(['LogSalePrice','Id','SalePrice'],axis =1)
Y = df_le['LogSalePrice']
```

### Gradient Boosting Model

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor

gbm = GradientBoostingRegressor()
gbm.fit(X,Y)
gbm.score(X,Y)
```

```python
gbm.set_params(verbose=1)
gbm.fit(X,Y) 
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
# a train test split of 20% test set
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2)
```

#### Bagging

```python
gbm.set_params(learning_rate=0.01, n_estimators=600, verbose=1, subsample=0.7, warm_start=False, loss='ls', max_depth=2, max_features=2, min_impurity_decrease=0.01)
# subsample = 0.7: each tree only use random 70% of the whole dataset for bagging
```

```python
gbm.fit(X_train,Y_train)
print('-'*50)
print("The train set R^2 is %.3f" %(gbm.score(X_train, Y_train)))
print("The test set R^2 is %.3f" %(gbm.score(X_test,Y_test)))
```

```python
sorted_importance = sorted(zip(X.columns.to_numpy(), gbm.feature_importances_), key=lambda t:t[1], reverse=True)
sorted_importance_top50 = sorted_importance[:50]
sorted_importance_top50
```

```python
a, b = zip(*sorted_importance_top50) # this splits the list of tuples into two lists
pd.DataFrame({'feature_name':a, 'importance_score':b}).plot.bar(x='feature_name',y='importance_score', legend=None)
plt.title('Feature Importance Plot of 500-Tree GBM')
plt.rcParams['figure.figsize'] = (10,15)
```

### Grid Search

```python
from sklearn.model_selection import GridSearchCV
grid_para_tree = [{
    'n_estimators':np.arange(100,1000,100),
    'learning_rate':np.linspace(0.01,0.1,3),
    "min_samples_leaf": range(1, 10),
    "min_samples_split": np.linspace(start=2, stop=30, num=7, dtype=int) # not recommend to do a through research, try num =7 first 
}]

gbm.set_params(random_state=108)
grid_search_tree = GridSearchCV(gbm, grid_para_tree, cv=3, n_jobs=2)
%time grid_search_tree.fit(X_train, Y_train)
```
